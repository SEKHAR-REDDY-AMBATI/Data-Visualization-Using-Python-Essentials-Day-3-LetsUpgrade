```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
from numpy.random import randn
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', data = fmri , hue = 'region', style = 'event', size = 'subject')
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri , hue = 'region', style = 'event', size = 'subject')
```

```python
sns.boxplot(x = 'timepoint', y = 'signal', data = fmri , hue = 'region')
```

```python
sns.catplot(x = 'timepoint', y = 'signal', data = fmri , hue = 'region',kind="box")
```

```python
sns.catplot(x = 'timepoint', y = 'signal', data = fmri , hue = 'region',kind="boxen")
```
